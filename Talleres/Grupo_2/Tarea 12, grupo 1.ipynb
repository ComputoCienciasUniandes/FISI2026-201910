# Punto 1 - Grupo 1

Hallar el área bajo la curva en esta función

$f(x,y) = \cos(\pi x) e^{-(x^2 + y^2) / (2 \pi ^2)} + 5$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from scipy.optimize import curve_fit
from scipy.integrate import dblquad
```

```python
def f(x,y):
    return np.cos(np.pi * x) * np.exp( -( x**2 + y**2)/(2*np.pi**2)) + 5
```

# Punto 2 - Muestro con el algoritmo Metrópolis-Hastings

La variable x está definida dentro del dominio real (0, 20). Se tiene una
distribución de probabilidad no normalizada f (x) definida como:

\begin{equation}
 f(x) = \begin{cases} 2, \textrm{si } x \in (0,4) \\
          4, \textrm{si } x \in (4,10) \\
          1, \textrm{si } x \in (10,20) \end{cases}
\end{equation}

Implemente el método de Metropolis-Hastings para hacer un muestreo de
N = 10000 puntos.


Se genera un caminante aleatorio (Random Walk) que da cada paso según la
función f (x) y la posición actual del caminante (Cadena de Markov). El decidir
si se da el siguiente paso aleatorio o si se permanece en la posición actual se
realiza según el algoritmo de Metropolis y Hastings.


```python
def f(x):
    a = 4
    b = 10
    c = 20
    
    x = x*1.
    
    if x < 0:
        return -np.inf
    if x < a:
        return 2
    if x < b:
        return 4
    if x < c:
        return 1
    if x >= 20:
        return -np.inf
```

```python
def ProbMonteCarlo(N, Deltax):
    walk = []
    x = np.random.rand()*20

    walk.append(x)

    for i in range(N):

        dx = Deltax*2*(np.random.rand() -0.5)

        x_o = walk[-1]

        f_o = f( x_o)
        f_n = f(dx + x_o)

        alpha = f_n / f_o

        if f_n == -np.inf:
            walk.append(x_o)
        else:
            if alpha >= 1.0:
                walk.append(dx + x_o)
            else:
                beta = np.random.rand()
                if beta <= alpha:
                    walk.append(x_o + dx)
                else:
                    walk.append(x_o)

    y, bins, algo = plt.hist(walk, bins=np.linspace(-2,22,51))
```

```python
ProbMonteCarlo(10000,5)
```

```python
ProbMonteCarlo(10000,100)
```
